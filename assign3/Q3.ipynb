```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('universities_std.csv')
print(df)
```

```python
df.drop(['Unnamed: 0','Top10','Accept','SFRatio','Expenses'],axis=1,inplace=True)
df.columns
```

```python
no_of_clusters=np.arange(1,11)

wss=[]

from sklearn.cluster import KMeans

for k in no_of_clusters:
    model=KMeans(n_clusters=k,random_state=12345)
    model.fit(df)
    wss.append(model.inertia_)





plt.plot(no_of_clusters, wss, label="WSS")
plt.scatter(no_of_clusters, wss)


plt.scatter(no_of_clusters[2], wss[2],
            color="red", label="Optimal K", marker="s")
plt.xlabel("No of clusters")
plt.ylabel("Within Sum of Squares")
plt.title("Elbow Method Visualization")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3, random_state=12345)
model.fit(df)


colors = ['red', 'green', 'blue']
markers = ['o', 'v', '^']

plt.xlabel("SAT")
plt.ylabel("GradRate")
plt.title("Clustering Universities_std")

for index in range(len(model.cluster_centers_)):
    plt.scatter(
        df['SAT'][model.labels_ == index],
        df['GradRate'][model.labels_ == index],
        color=colors[index], marker=markers[index])
    plt.scatter(
        model.cluster_centers_[index][0],
        model.cluster_centers_[index][1],
        color="black", marker="s")

plt.legend()
plt.show()
```
