```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('East_std.csv')
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
print(df)
```

```python
# number of clusters
no_of_clusters = np.arange(1, 11)

# collect all WSS (inertia)
wss = []

from sklearn.cluster import KMeans

for k in no_of_clusters:
    # create the model for k number of clusters
    model = KMeans(n_clusters=k, random_state=12345)

    # fit the data
    model.fit(df)

    # collect the wss for elbow method
    wss.append(model.inertia_)
    print(wss)


# plot a chart for Elbow method
plt.plot(no_of_clusters, wss, label="WSS")
plt.scatter(no_of_clusters, wss)
plt.scatter(no_of_clusters[4], wss[4],
            color="red", label="Optimal K", marker="s")
plt.xlabel("No of clusters")
plt.ylabel("Within Sum of Squares")
plt.title("Elbow Method Visualization")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=4, random_state=12345)
model.fit(df)
print(model.labels_)
print(model.cluster_centers_)
```

```python
plt.xlabel("Balance")
plt.ylabel("Qual_miles")
plt.title("Clustering based on miles")

plt.scatter(
    df['Balance'][model.labels_ == 0],
    df['Qual_miles'][model.labels_ == 0],
    color="red", marker='o')
plt.scatter(
    model.cluster_centers_[0][0],
    model.cluster_centers_[0][1],
    color="black", marker="s")
```

```python
colors = ['red', 'green', 'blue', 'brown', 'magenta','yellow','orange','cyan','grey','silver','pink']
markers = ['o', 'v', '^', 'x', '+','o', 'v', '^', 'x', '+','o']

plt.xlabel("Balance")
plt.ylabel("Qual_miles")
plt.title("Clustering Customers")

for index in range(len(model.cluster_centers_)):
    plt.scatter(
        df['Balance'][model.labels_ == index],
        df['Qual_miles'][model.labels_ == index],
        color=colors[index], marker=markers[index])
    plt.scatter(
        model.cluster_centers_[index][0],
        model.cluster_centers_[index][1],
        color="black", marker="s")
```
