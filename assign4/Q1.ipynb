```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Hierarchical Clustering

```python
df = pd.read_csv('East_std.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['Unnamed: 0','cc1_miles','cc2_miles','cc3_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','Flight_trans_12','Days_since_enroll','Award?'
],axis=1,inplace=True)
df.columns
```

```python
from scipy.cluster import hierarchy
```

```python
linkage = hierarchy.linkage(df, 'complete')


clusters_info = hierarchy.dendrogram(linkage)
```

```python
from sklearn.cluster import AgglomerativeClustering


model = AgglomerativeClustering(n_clusters=5, linkage='complete')


model.fit(df)
```

```python
model.labels_
```

```python
figure = plt.figure()


ax = figure.add_subplot(111, projection='3d')

colors = ['red', 'green', 'blue', 'magenta', 'yellow']
for index in range(len(colors)):
    ax.scatter(df['Balance'][model.labels_ == index], df['Qual_miles'][model.labels_ == index], color=colors[index])

plt.xlabel('Balance')
plt.ylabel('Qual_miles')
```

## DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN

# create an object of DBScan
model = DBSCAN(eps=0.015, min_samples=8)

# fit the data
model.fit(df)
```

```python
model.labels_
```

```python
figure = plt.figure()

# add a subplot
ax = figure.add_subplot(111)

colors = ['red', 'green', 'blue', 'magenta', 'yellow', "brown"]
for index in range(len(colors)):
    ax.scatter(df['Balance'][model.labels_ == index], df['Qual_miles'][model.labels_ == index], color=colors[index])

plt.xlabel('Balance')
plt.ylabel('Qual_miles')
```
