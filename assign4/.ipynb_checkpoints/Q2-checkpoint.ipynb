```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('crime_std.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(["Unnamed: 0", "Murder", "Assault"], axis=1, inplace=True)
```

```python
df.columns
```

```python
from scipy.cluster import hierarchy
```

```python
linkage = hierarchy.linkage(df, 'complete')
clusters_info = hierarchy.dendrogram(linkage)
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=5, linkage='complete')


model.fit(df)
```

```python
figure = plt.figure()

# add a subplot
ax = figure.add_subplot(111, projection='3d')

colors = ['red', 'green', 'blue', 'magenta', 'yellow']
for index in range(len(colors)):
    ax.scatter(df['UrbanPop'][model.labels_ == index], df['Rape'][model.labels_ == index], color=colors[index])

plt.xlabel('UrbanPop')
plt.ylabel('Rape')
```

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.2, min_samples=5)

model.fit(df)
```

```python
model.labels_
```

```python
figure = plt.figure()

# add a subplot
ax = figure.add_subplot(111)

colors = ['red', 'green', 'blue', 'magenta', 'yellow', "brown"]
for index in range(len(colors)):
    ax.scatter(df['UrbanPop'][model.labels_ == index], df['Rape'][model.labels_ == index], color=colors[index])

plt.xlabel('UrbanPop')
plt.ylabel('Rape')
```
