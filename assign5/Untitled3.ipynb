## In the given dataset TVMarketing.csv you have given  sales based on 'TV' marketing budget.
## Build a linear regression model to predict 'Sales' using 'TV' as the predictor variable. Also plot regression line.

## Dataset : TVMarketing.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TvMarketing.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
y = df['Sales']
x = df.drop('Sales' ,axis=1)
```

```python
from sklearn.linear_model import LinearRegression

# create a new object of LinearRegression
model = LinearRegression()

# fit the data
model.fit(x, y)
```

```python
mean_values = np.ones(y.size) * y.mean()
mean_values
```

```python
plt.scatter(x['TV'], y, label='Observed Values', color="green") 

# draw mean line
mean_values = np.ones(y.size) * y.mean()
plt.plot(x['TV'], mean_values, color='red')
plt.scatter(x['TV'], mean_values, color='red', label="mean values")

# draw the regression line
plt.plot(x['TV'], model.predict(x), color='blue')
plt.scatter(x['TV'], model.predict(x), color='blue', label="expected values")


# title
plt.title('Regression Line graph')
plt.xlabel('TV')
plt.ylabel('Sales' )

plt.legend()
```
