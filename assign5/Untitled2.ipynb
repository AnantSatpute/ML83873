## delivery_time.csv
## Independent = ‘Sorting Time’, Dependent=’Delivery Time’

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('delivery_time.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
y = df['Delivery Time']
x = df.drop('Delivery Time',axis=1)
```

```python
from sklearn.linear_model import LinearRegression

# create a new object of LinearRegression
model = LinearRegression()

# fit the data
model.fit(x, y)
```

```python
mean_values = np.ones(y.size) * y.mean()
mean_values
```

```python
plt.scatter(x['Sorting Time'], y, label='Observed Values', color="green") 

# draw mean line
mean_values = np.ones(y.size) * y.mean()
plt.plot(x['Sorting Time'], mean_values, color='red')
plt.scatter(x['Sorting Time'], mean_values, color='red', label="mean values")

# draw the regression line
plt.plot(x['Sorting Time'], model.predict(x), color='blue')
plt.scatter(x['Sorting Time'], model.predict(x), color='blue', label="expected values")


# title
plt.title('Regression Line graph')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')

plt.legend()
```
