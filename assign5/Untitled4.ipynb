## You own an ice cream business and you would like to create a model that could predict the daily revenue in dollars based on the outside air temperature (degC). Build Linear Regression Model and plot regression line.

## Dataset : IceCreamData.csv


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IceCreamData.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
y = df['Revenue']
x = df.drop('Revenue',axis=1)
```

```python
from sklearn.linear_model import LinearRegression

# create a new object of LinearRegression
model = LinearRegression()

# fit the data
model.fit(x, y)
```

```python
mean_values = np.ones(y.size) * y.mean()
mean_values
```

```python
plt.scatter(x['Temperature'], y, label='Observed Values', color="green") 

# draw mean line
mean_values = np.ones(y.size) * y.mean()
plt.plot(x['Temperature'], mean_values, color='red')
plt.scatter(x['Temperature'], mean_values, color='red', label="mean values")

# draw the regression line
plt.plot(x['Temperature'], model.predict(x), color='blue')
plt.scatter(x['Temperature'], model.predict(x), color='blue', label="expected values")


# title
plt.title('Regression Line graph')
plt.xlabel('Temperature')
plt.ylabel('Revenue')

plt.legend()
```
