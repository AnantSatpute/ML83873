## In the following dataset you have given SAT score and GPA of students. You have to predict GPA of students by using SAT score.
## Build a Linear Regression Model and plot regression line.

## Dataset : SAT-GPA.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SAT-GPA.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
y = df['GPA']
x = df.drop('GPA',axis=1)
```

```python
from sklearn.linear_model import LinearRegression

# create a new object of LinearRegression
model = LinearRegression()

# fit the data
model.fit(x, y)
```

```python
mean_values = np.ones(y.size) * y.mean()
mean_values
```

```python
plt.scatter(x['SAT'], y, label='Observed Values', color="green") 

# draw mean line
mean_values = np.ones(y.size) * y.mean()
plt.plot(x['SAT'], mean_values, color='red')
plt.scatter(x['SAT'], mean_values, color='red', label="mean values")

# draw the regression line
plt.plot(x['SAT'], model.predict(x), color='blue')
plt.scatter(x['SAT'], model.predict(x), color='blue', label="expected values")


# title
plt.title('Regression Line graph')
plt.xlabel('SAT')
plt.ylabel('GPA')

plt.legend()
```
